```python
import yaml
import numpy as np
```

```python
def generate():
    yaml_dict = {}
    array = np.arange(0.1,.5,0.1)
    array_2 = np.arange(0.5,1,0.2)
    array_3 = np.arange(1,2,0.3)
    array_4 = np.arange(0.1,4.1,0.02)
    wavelengths=np.arange(400,1075,25)
    fiber_spectrograph_list = []
    #slit_list = array.tolist() + array_2.tolist() + array_3.tolist() + array_4.tolist()
    slit_list = array_4.tolist()
    #slit_list = [0.1,0.2,0.4,0.7,1.0,1.5,2.0,2.5,3.0,4.0]
    yaml_dict["mono_exit_slit_widths"]=slit_list
    yaml_dict["mono_entrance_slit_widths"]=slit_list
    yaml_dict["wavelengths"]=500
    yaml_dict["grating_types"]=0
    #yaml_dict["mono_exit_slit_widths"]=1
    #yaml_dict["mono_entrance_slit_width"]=1
    
    for slits in slit_list:
        if slits < 0.5:
            fiber_spectrograph_list.append(0.2)
            continue
        if slits < 1.0:
            fiber_spectrograph_list.append(0.1)
            continue
        if slits < 2.0:
            fiber_spectrograph_list.append(0.05)
            continue
        if slits < 2.5:
            fiber_spectrograph_list.append(0.01)
            continue
        else:
            fiber_spectrograph_list.append(0.01)
            continue
        
    yaml_dict["fiber_spectrograph_integration_times"]=fiber_spectrograph_list
    print(len(fiber_spectrograph_list))
    print(len(slit_list))
    return yaml.dump(yaml_dict)
```
