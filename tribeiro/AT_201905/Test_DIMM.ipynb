```python
import asyncio

from lsst.ts import salobj

import SALPY_DIMM
```

```python
r = salobj.Remote(SALPY_DIMM, 1)
```

```python
await r.evt_heartbeat.next(flush=True, timeout=10)
```

```python
state = salobj.State(r.evt_summaryState.get().summaryState)
print(f"{r.salinfo.name}::{state!r}")
```

```python
r.cmd_start.set(settingsToApply="lsst")
await r.cmd_start.start()
```

```python
await r.cmd_enable.start()
```

```python
await salobj.set_summary_state(r, salobj.State.ENABLED, settingsToApply="lsst")
```

```python
while True:
    seeing =  await r.evt_dimmMeasurement.next(flush=False)
    print(seeing.fwhm)
```

```python
await salobj.set_summary_state(r, salobj.State.STANDBY)
```
